```python
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv('kag_risk_factors_cervical_cancer.csv')

# replace missing values with NaN
df = df.replace('?', np.nan)

# impute missing values with mode
imputer = SimpleImputer(strategy='most_frequent')
imputed_df = pd.DataFrame(imputer.fit_transform(df))
imputed_df.columns = df.columns
imputed_df.index = df.index


```

```python


def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df



for column in ["Age","Number of sexual partners","First sexual intercourse","Num of pregnancies", "Smokes", "Smokes (years)", "Smokes (packs/year)","Hormonal Contraceptives","Hormonal Contraceptives (years)","IUD","IUD (years)"]:
    imputed_df= remove_outliers(imputed_df, column)



```

```python
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA

# # separate features and target variables
X = imputed_df.drop(columns=['Hinselmann', 'Schiller', 'Citology', 'Biopsy'])
y_hinselmann = imputed_df['Hinselmann']
y_schiller = imputed_df['Schiller']
y_citology = imputed_df['Citology']
y_biopsy = imputed_df['Biopsy']
# # perform PCA

# data_pca = pca.fit(X)
column = ['Smokes','Smokes (years)','Smokes (packs/year)','IUD','IUD (years)','STDs:cervical condylomatosis','STDs:AIDS','STDs:Hepatitis B','STDs:HPV']
X  = X.drop(column,axis=1)

X= (X-X.mean())/X.std()
pca = PCA()
pca.fit(X)


relevance = pca.explained_variance_ratio_
relevant_features = pd.DataFrame({'feature': X.columns, 'importance': relevance})
relevant_features = relevant_features.sort_values('importance', ascending=False)

top_columns = list(relevant_features['feature'][:12])
X2 =  imputed_df[top_columns] # data_FOR_1ST is the data which is containe data after applying feature extraction 
X2.head()
X2.head(20)
```

```python
sns.heatmap(X2.corr(), annot=False)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# slice y_biopsy to have the same number of samples as X_pca
y_biopsy_sliced = y_biopsy[:X2.shape[0]]
y_hinselmann_sliced = y_hinselmann[:X2.shape[0]]
y_citology_sliced = y_citology[:X2.shape[0]]
y_schiller_sliced = y_schiller[:X2.shape[0]]

# Split data into training and testing sets for the four target variables
X_train1, X_test1, y_train1, y_test1 = train_test_split(X2,y_hinselmann_sliced, test_size=0.3, random_state=42, stratify=y_hinselmann_sliced)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y_schiller_sliced, test_size=0.3, random_state=42, stratify=y_schiller_sliced)
X_train3, X_test3, y_train3, y_test3 = train_test_split(X2, y_citology_sliced, test_size=0.3, random_state=42, stratify=y_citology_sliced)
X_train4, X_test4, y_train4, y_test4 = train_test_split(X2, y_biopsy_sliced, test_size=0.3, random_state=42, stratify=y_biopsy_sliced)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train_balanced1, y_train_balanced1 = sm.fit_resample(X_train1, y_train1)
X_train_balanced2, y_train_balanced2 = sm.fit_resample(X_train2, y_train2)
X_train_balanced3, y_train_balanced3 = sm.fit_resample(X_train3, y_train3)
X_train_balanced4, y_train_balanced4 = sm.fit_resample(X_train4, y_train4)


```

```python

from sklearn import svm
# SVM classifier for Hinselmann target variable
svm_hinselmann = svm.SVC(C=1, kernel='rbf', gamma='scale')
svm_hinselmann.fit(X_train_balanced1, y_train_balanced1)
y_pred_svm_hinselmann = svm_hinselmann.predict(X_test1)


svm_schiller = svm.SVC(C=1, kernel='rbf', gamma='scale')
svm_schiller.fit(X_train_balanced2, y_train_balanced2)
y_pred_svm_schiller = svm_schiller.predict(X_test2)


svm_citology = svm.SVC(C=1, kernel='rbf', gamma='scale')
svm_citology.fit(X_train_balanced3, y_train_balanced3)
y_pred_svm_citology = svm_citology.predict(X_test3)


svm_biopsy = svm.SVC(C=1, kernel='rbf', gamma='scale')
svm_biopsy.fit(X_train_balanced4, y_train_balanced4)
y_pred_svm_biopsy = svm_biopsy.predict(X_test4)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
print('Accuracy:', accuracy_score(y_test1, y_pred_svm_hinselmann))
print('Precision:', precision_score(y_test1, y_pred_svm_hinselmann))
print('Recall:', recall_score(y_test1, y_pred_svm_hinselmann))
sns.heatmap(confusion_matrix(y_test1, y_pred_svm_hinselmann), annot=True)
plt.show()
```

```python

print('Accuracy:', accuracy_score(y_test2, y_pred_svm_schiller))
print('Precision:', precision_score(y_test2, y_pred_svm_schiller))
print('Recall:', recall_score(y_test2, y_pred_svm_schiller))
sns.heatmap(confusion_matrix(y_test2, y_pred_svm_schiller), annot=True)
plt.show()
```

```python

print('Accuracy:', accuracy_score(y_test3, y_pred_svm_citology))
print('Precision:', precision_score(y_test3, y_pred_svm_citology))
print('Recall:', recall_score(y_test3, y_pred_svm_citology))
sns.heatmap(confusion_matrix(y_test3, y_pred_svm_citology), annot=True)
plt.show()
```

```python

print('Accuracy:', accuracy_score(y_test4, y_pred_svm_biopsy))
print('Precision:', precision_score(y_test4, y_pred_svm_biopsy))
print('Recall:', recall_score(y_test4, y_pred_svm_biopsy))
sns.heatmap(confusion_matrix(y_test4, y_pred_svm_biopsy), annot=True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# KNN classifier for Hinselmann target variable
knn_hinselmann = KNeighborsClassifier(n_neighbors=5)
knn_hinselmann.fit(X_train_balanced1, y_train_balanced1)
y_pred_knn_hinselmann = knn_hinselmann.predict(X_test1)

knn_schiller = KNeighborsClassifier(n_neighbors=5)
knn_schiller.fit(X_train_balanced2, y_train_balanced2)
y_pred_knn_schiller = knn_schiller.predict(X_test2)

knn_citology = KNeighborsClassifier(n_neighbors=5)
knn_citology.fit(X_train_balanced3, y_train_balanced3)
y_pred_knn_citology = knn_citology.predict(X_test3)

knn_biopsy = KNeighborsClassifier(n_neighbors=5)
knn_biopsy.fit(X_train_balanced4, y_train_balanced4)
y_pred_knn_biopsy = knn_biopsy.predict(X_test4)
```
